```python
import pandas as pd
import numpy as np
import hvplot.pandas  # noqa
hvplot.extension('plotly')
```

```python
df = pd.read_csv("./Tesla Stock Price (2010 to 2023).csv", sep=",")


```

```python
import plotly.graph_objects as go

plot = go.Figure(go.Scatter(x=df.Date, y=df.Open))
plot.update_layout(
    xaxis_tickformat = '%d %B (%a)<br>%Y'
)
plot.show()
```
